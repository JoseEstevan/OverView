```python
import pandas as pd
import numpy as np
import altair as alt
import matplotlib.pyplot as plt
pd.set_option('plotting.backend', 'pandas_bokeh')
df = pd.read_csv('datahackers-survey.csv')
df.columns
df['Estado'] = df["('P5', 'living_state')"]
df['Formação'] = df["('P8', 'degreee_level')"]
df['Linguagem de Programação'] = df["('P22', 'most_used_proggraming_languages')"]
df['Trabalho'] = df["('P10', 'job_situation')"]

```

```python
df['Salário Médio'] = df["('P16', 'salary_range')"].fillna('$ 0/').apply(lambda x: #get the mean in ranges
                    int( #transform all in int in the end
                    (int(str(x)[str(x).rfind(' ')+1:str(x).rfind('/')].replace('.','')) # Get max in range 
                    +
                    int(str(x)[str(x).find('$')+2:str(x).find('/')].replace('.','')) #Get min
                    )/2)) #divide by 2
```

```python
col_xp = "('P18', 'time_experience_before')"
df.loc[df[col_xp] == 'Não tive experiência na área de TI/Engenharia de Software antes de começar a trabalhar na área de dados', col_xp] = 'Sem xp em TI antes'
```

```python
import matplotlib.pyplot as plt
df['Estado'].value_counts().plot.barh(color='#16003d', title='Regiões com mais profissionais na área de dados', xlabel='Número de pessoas')
```

```python
df['Trabalho'].value_counts().plot.barh(color='#D81B71', title='Situação de Trabalho', xlabel='Número de pessoas')
```

```python
df['Linguagem de Programação'].value_counts().nlargest(4).plot.bar(color='#17A87C', title='Linguagens de Programação mais usadas', ylabel='Número de pessoas', xlabel='Linguagens')
```

```python
df['Formação'].value_counts().plot.barh(color='#35C2C9', title='Formação Acadêmica', xlabel='Número de pessoas')
```

```python
df.groupby("('P10', 'job_situation')")['Salário Médio'].mean().nlargest(6).plot.barh(color='#3640BF', title='Situação X Salário', ylabel='Situação')
```

```python
df.groupby('Estado')['Salário Médio'].mean().nlargest(6).plot.barh(color='#0D33F1', title='Estado X Salário', ylabel='')
```

```python
df.groupby("('P18', 'time_experience_before')")['Salário Médio'].mean().plot.barh(color='#179A19', title='Experiência X Salário', xlabel='Salário Médio', ylabel='Experiência') 
```
